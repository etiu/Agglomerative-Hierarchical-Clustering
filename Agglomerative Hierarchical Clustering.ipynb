```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df
```

```python
df.describe()
```

```python
df1 = df.drop('Id', axis = 1)
```

```python
df1
```

```python
import scipy
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X = df1[['SepalLengthCm','PetalLengthCm']].values.reshape(-1,2)
scipy.spatial.distance.pdist(X, metric='euclidean')
plt.title("Single linkage based Dendrograms")  
dend = dendrogram(linkage(X, method='single'))
```

```python
X = df1[['SepalLengthCm','PetalLengthCm']].values.reshape(-1,2)
scipy.spatial.distance.pdist(X, metric='cityblock')
plt.title("Single linkage based Dendrograms")  
dend = dendrogram(linkage(X, method='single'))
```

```python
X = df1[['SepalLengthCm','PetalLengthCm']].values.reshape(-1,2)
scipy.spatial.distance.pdist(X, metric='euclidean')
plt.title("Complete linkage based Dendrograms")  
dend = dendrogram(linkage(X, method='complete'))
```

```python
X = df1[['SepalLengthCm','PetalLengthCm']].values.reshape(-1,2)
scipy.spatial.distance.pdist(X, metric='cityblock')
plt.title("Complete linkage based Dendrograms")  
dend = dendrogram(linkage(X, method='complete'))
```

```python
K = 5
L = 'complete'
```

```python
cluster = AgglomerativeClustering(n_clusters=K, affinity='euclidean', linkage=L)
pred = cluster.fit_predict(X)
pred
plt.scatter(X[:,0], X[:,1], c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(X, pred)
```
